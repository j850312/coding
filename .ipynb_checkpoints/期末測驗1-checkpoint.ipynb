```python
import re
f=open("F:\\陽明coding\\python\\4.KT388711.1.GB","r")
data=f.readline()
while data:
    check=re.search(r"LOCUS",data)
    if(check!=None):
        ans=""
        tp=data.split("VRL ")
        tp1=tp[1].split("\n")
        name=data.split("       ")
        ans+=name[1]+" "+tp1[0]+" "
    check1=re.search(r"/organism",data)
    if(check1!=None):
        temp=data.split("\"")
        if(re.search("\n",temp[1])!=None):
            temp=temp[1].split("\n")
            ans+=temp[0]
        else:
            ans+=temp[1]+" "
    check2=re.search(r"/strain",data)
    if(check2!=None):
        temp1=data.split("\"")

        ans+=temp1[1]
    check3=re.search(r"/serotype",data)
    if(check3!=None):
        temp2=data.split("\"")
        ans+="("+temp2[1]+")"
        print(ans)
    data=f.readline()
```
